### Task 1: NumPy Basics
Write a Python script that creates a NumPy array with random integers and performs basic mathematical operations like addition, subtraction, multiplication, and division on it.


```python
import numpy as np

array = 5
minV = 1
maxV = 100
random_array = np.random.randint(minV, maxV + 1, size=array)

print("Random Array:")
print(random_array)

num = 10

add = random_array + num
sub = random_array - num
mul = random_array * num
div = random_array / num

print("\nAddition Result:")
print(add)

print("\nSubtraction Result:")
print(sub)

print("\nMultiplication Result:")
print(mul)

print("\nDivision Result:")
print(div)
```

### Task 4: NumPy Array Manipulation
Create two NumPy arrays and perform operations to concatenate them vertically and horizontally.


```python
!pip install matplotlib
```

```python
import numpy as np

# Create two NumPy arrays
array1 = np.array([[1, 2], [3, 4]])
array2 = np.array([[5, 6]])

# Vertically concatenate them
ver = np.vstack((array1, array2))

# Horizontally concatenate them
hor = np.hstack((array1, array2.T))  # Transpose array2 to match dimensions

print("Array 1:")
print(array1)

print("\nArray 2:")
print(array2)

print("\nVertically Concatenated:")
print(ver)

print("\nHorizontally Concatenated:")
print(hor)
```

### Task 6: Matplotlib Customization
Create a line plot or scatter plot using Matplotlib and customize it by adding a legend, different marker styles, and changing the line colors


```python
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(8, 6))

plt.plot(x, y1, label='sin(x)', color='blue', linestyle='-', marker='o', markersize=5)

plt.plot(x, y2, label='cos(x)', color='red', linestyle='--', marker='s', markersize=5)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Customized Line Plot')
plt.legend()

plt.grid(True)
plt.show()
```

### Task 7: NumPy Statistical Analysis
Use NumPy to calculate descriptive statistics like variance, standard deviation, and correlation coefficients for a dataset of your choice (you can find datasets online or use your own).


```python
import numpy as np

X = np.array(df['age'])
Y = np.array(df['height'])

variance_X = np.var(X)
deviation_X = np.std(X)
variance_Y = np.var(Y)
deviation_Y = np.std(Y)

correlation_coefficient = np.corrcoef(X, Y)[0, 1]


print("Descriptive Statistics for X:")
print(f"Variance: {variance_X}")
print(f"Standard Deviation: {deviation_X}")

print("\nDescriptive Statistics for Y:")
print(f"Variance: {variance_Y}")
print(f"Standard Deviation: {deviation_Y}")

print("\nCorrelation Coefficient between X and Y:")
print(f"Correlation Coefficient: {correlation_coefficient}")
```
